```python
#import necessary libraries
import keras 
```

```python
import pandas as pd
```

```python
df=pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
df.head()#reading the data into a pandas dataframe
```

```python
#normalising data
df=(df-df.mean())/df.std()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
#importing train_test_splir
```

```python
l=[]
```

```python
#For loop for randomly splitting the dataset and building and compiling the model
#ading MSE to list l
for i in range(1,51):
    train_x,test_x,train_y,test_y=train_test_split(df[['Cement', 'Blast Furnace Slag', 'Fly Ash', 'Water',
       'Superplasticizer', 'Coarse Aggregate', 'Fine Aggregate', 'Age']],df['Strength'],test_size=0.3)
    model=keras.Sequential([keras.layers.Dense(128,activation="relu",input_shape=(train_x.shape[1],)),
                           keras.layers.Dense(256,activation="relu"),
                           keras.layers.Dense(1)])
    model.compile(optimizer="adam",loss="mean_squared_error")
    model.fit(train_x,train_y,epochs=50)
    l.append(model.evaluate(test_x,test_y))
print(l)
```

```python
import numpy as np
sum=0
#for loop to print the MSE with mean and standard deviation
for i in range(0,50):
    sum=sum+l[i]
    print("MSE for iteration "+str(i+1)+"="+str(l[i]))
print("mean="+str(sum/50))
print("std="+str(np.std(l)))
```
